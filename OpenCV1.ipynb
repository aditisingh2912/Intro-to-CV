Reading, displaying, and writing an image using OpenCV.

```python
import cv2
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/sato.png")
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imwrite("output.jpg", image)

```

```python
cv2.imwrite("output.png", image)
```

```python
resized_image = cv2.resize(image, (300, 300))
cv2_imshow(resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
from google.colab.patches import cv2_imshow
cv2_imshow(gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
scaled_image = cv2.resize(image, None, fx=1.5, fy=1.5)

(h, w) = image.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

shift_matrix = np.float32([[1, 0, 50], [0, 1, 50]])
shifted_image = cv2.warpAffine(image, shift_matrix, (w, h))


cv2_imshow(scaled_image)
cv2_imshow(rotated_image)
cv2_imshow(shifted_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture("video.mp4")

while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break
    cv2.imshow("Video", frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture("video.mp4")
frame_count = 0

while video.isOpened():
    ret, frame = video.read()
    if not ret:
        break

    cv2.imwrite(f"frame_{frame_count}.jpg", frame)
    frame_count += 1

video.release()
```

```python
import numpy as np
import matplotlib.pyplot as plt
f = float(input("Enter frequency: "))
rows, cols = 256, 256
image = np.zeros((rows, cols))

for i in range(rows):
    for j in range(cols):
        image[i, j] = np.sin(2 * np.pi * f * (i + j))
image = (image - np.min(image)) / (np.max(image) - np.min(image)) * 255
plt.imshow(image, cmap="gray")
plt.title(f"Generated Image with Frequency {f}")
plt.show()
```
